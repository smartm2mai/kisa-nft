```python
#Training Autoencoder model

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tqdm 


image_size = 224
batch_size = 32
seed = 42

# Define the encoder
encoder = tf.keras.Sequential([
    layers.Input(shape=(image_size, image_size, 3)),
    layers.Conv2D(64, kernel_size=3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, kernel_size=3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(16, kernel_size=3, padding='same', activation='relu'),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
])

# Define the decoder
decoder = tf.keras.Sequential([
    layers.Dense(16*56*56, activation='relu'),  # Assuming the encoder ends with feature maps of size 56x56
    layers.Reshape((56, 56, 16)),
    layers.Conv2DTranspose(32, kernel_size=3, strides=2, padding='same', activation='relu'),  # Upsampling
    layers.Conv2DTranspose(64, kernel_size=3, strides=2, padding='same', activation='relu'),  # Upsampling
    layers.Conv2D(3, kernel_size=3, padding='same', activation='sigmoid'),  # Final layer to get back to input shape
])

# Combine the encoder and decoder into an autoencoder model
autoencoder = tf.keras.Model(inputs=encoder.input, outputs=decoder(encoder.output))

autoencoder.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss='mse')  # Use Mean Squared Error as the loss

# Image loading and augmentation
datagen = ImageDataGenerator(rescale=1./255)

train_generator = datagen.flow_from_directory(
    '../datasets/train',
    target_size=(image_size, image_size),
    batch_size=batch_size,
    class_mode='input',
    seed=seed)

val_generator = datagen.flow_from_directory(
    '../datasets/test',
    target_size=(image_size, image_size),
    batch_size=batch_size,
    class_mode='input',
    seed=seed)

model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    './model/autoencoder_best_val.h5',
    monitor="val_loss",
    verbose=1,
    save_best_only=True,
    save_freq="epoch",
)

history = autoencoder.fit(
    train_generator,
    steps_per_epoch = train_generator.samples // batch_size,
    validation_data = val_generator,
    validation_steps = val_generator.samples // batch_size,
    epochs = 20,
    callbacks=[model_checkpoint_callback])

# Plotting
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
#Test to take 5 test sample image to be checked/verified 
# Setting the threshold 0.05 (approve if the score less than 0.05, reject if the score more than 0.05)

import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model

def load_images_from_directory(start_id, end_id, base_dir):
    image_ids = []
    image_arrays = []
    for i in range(start_id, end_id+1):
        img_dir = os.path.join(base_dir, str(i))
        for filename in os.listdir(img_dir):
            img_path = os.path.join(img_dir, filename)
            img = load_img(img_path, target_size=(image_size, image_size))
            img_array = img_to_array(img) / 255.0
            image_ids.append(f"{i}/{filename}")
            image_arrays.append(img_array)
            if len(image_ids) >= 5:  # Break the loop after loading 5 images
                break
        if len(image_ids) >= 5:  # Break the outer loop as well
            break
    return image_ids, image_arrays

def simulate_nft_upload(nft_id, nft_image, simulation_database, autoencoder, threshold=0.05):
    # Check if the NFT has already been uploaded
    if nft_id in simulation_database:
        return False, simulation_database

    # Compute the difference between the input and reconstructed image
    reconstructed_image = autoencoder.predict(np.expand_dims(nft_image, axis=0))
    difference = np.mean(np.abs(nft_image - reconstructed_image[0]))

    # Determine whether to approve the upload based on the difference and threshold
    is_approved = difference < threshold

    # Update the simulation database
    simulation_database[nft_id] = is_approved

    return is_approved, difference, simulation_database

# Load the trained autoencoder model
autoencoder = load_model('./model/autoencoder_best_val.h5')

# Define the image size for the loader
image_size = 224

# Load images from the test set
test_dir = "../datasets/test"
sample_nfts_ids, sample_nfts_images = load_images_from_directory(7001, 10000, test_dir)

# Initialize the simulation database
simulation_database = {}

# Simulate the upload of the sample NFTs and display the results
for nft_id, nft_image in zip(sample_nfts_ids, sample_nfts_images):
    is_approved, difference, simulation_database = simulate_nft_upload(nft_id, nft_image, simulation_database, autoencoder)
    # Display the image, the difference score and the result
    plt.imshow(nft_image)
    plt.title(f"Image '{nft_id}' was approved: {is_approved}, Score: {difference}")
    plt.show()
```
